```python
def download(filename, url):
    # We do not download again if the file already exists
    if os.path.exists(filename):
        return (str(filename) + " already exists!")
    
    get = requests.get(url)
    get.raise_for_status()
    data = get.text
    f = open(filename, "w", encoding="utf-8")
    f.write(data)
    f.close()

    # TODO: Write the code to download the file from URL
    # and save it in `filename`

    return (str(filename) + " created!")
```

```python
def distance(lat1, long1, lat2, long2):
    R = 3956
    dlon = math.radians(long2 - long1)
    dlat = math.radians(lat2 - lat1)
    a = ((math.sin(dlat/2))**2) + math.cos(math.radians(lat1)) * math.cos(math.radians(lat2)) * ((math.sin(dlon/2))**2)
    c = 2 * math.asin(min(1,math.sqrt(a)))
    d = R * c
    return d
```

```python
#Q13: What are the distances between Italy, United Kingdom and Germany?
country_names = ["Italy", "United Kingdom", "Germany"]
data_dict = {"Italy": [0, 0,0], "United Kingdom":[0, 0,0], "Germany":[0, 0,0]}

for key in data_dict: 
    for i in range(len(country_names)):
        if key == country_names[i]:
            data_dict[key][i] = None
        else:
            lat1 = next(iter(capitals[capitals["country"] == key]["latitude"]))
            long1 = next(iter(capitals[capitals["country"] == key]["longitude"]))
            lat2 = next(iter(capitals[capitals["country"] == country_names[i]]["latitude"]))
            long2 = next(iter(capitals[capitals["country"] == country_names[i]]["longitude"]))
            data_dict[key][i] = distance(lat1, long1, lat2, long2)

three_country_df = DataFrame(data_dict, columns=country_names, index=country_names)
three_country_df
```

```python
#Q14: What is the distance between every pair of countries in the North American continent?
country_list = countries[countries["continent"] == "North America"]["country"].tolist()
NA_data_dict = {}
for country in country_list:
    NA_data_dict[country] = []
    for i in range(len(country_list)):
        NA_data_dict[country].append(0)

for key in NA_data_dict: 
    for i in range(len(country_list)):
        if key == country_list[i]:
            NA_data_dict[key][i] = None
        else:
            lat1 = next(iter(capitals[capitals["country"] == key]["latitude"]))
            long1 = next(iter(capitals[capitals["country"] == key]["longitude"]))
            lat2 = next(iter(capitals[capitals["country"] == country_list[i]]["latitude"]))
            long2 = next(iter(capitals[capitals["country"] == country_list[i]]["longitude"]))
            NA_data_dict[key][i] = distance(lat1, long1, lat2, long2)

NA_country_df = DataFrame(NA_data_dict, columns=country_list, index=country_list)
NA_country_df
```

```python
#Q17: How close is each country in North America to its nearest neighbor?
NA_countries = NA_country_df.columns.tolist()
NA_nearest_dict = {"nearest":[], 'distance':[]}
for country in NA_countries:
    nearest_country = NA_country_df[country].idxmin()
    NA_nearest_dict["nearest"].append(nearest_country)

    #latitude and longitude of both countires
    lat1 = next(iter(capitals[capitals["country"] == country]["latitude"]))
    long1 = next(iter(capitals[capitals["country"] == country]["longitude"]))
    lat2 = next(iter(capitals[capitals["country"] == nearest_country]["latitude"]))
    long2 = next(iter(capitals[capitals["country"] == nearest_country]["longitude"]))
    dist = distance(lat1, long1, lat2, long2)
    NA_nearest_dict["distance"].append(dist)

NA_nearest_df = DataFrame(NA_nearest_dict, index=NA_countries)
NA_nearest_df
```

```python
#Q18: How far is each country in North America to its furthest neighbor?
NA_furthest_dict = {"furthest":[], 'distance':[]}
for country in NA_countries:
    furthest_country = NA_country_df[country].idxmax()
    NA_furthest_dict["furthest"].append(furthest_country)

    #latitude and longitude of both countires
    lat1 = next(iter(capitals[capitals["country"] == country]["latitude"]))
    long1 = next(iter(capitals[capitals["country"] == country]["longitude"]))
    lat2 = next(iter(capitals[capitals["country"] == furthest_country]["latitude"]))
    long2 = next(iter(capitals[capitals["country"] == furthest_country]["longitude"]))
    dist = distance(lat1, long1, lat2, long2)
    NA_furthest_dict["distance"].append(dist)

NA_furthest_df = DataFrame(NA_furthest_dict, index=NA_countries)
NA_furthest_df
```

```python
#Q20: BeautifulSoup
download('capitals.html', 'https://raw.githubusercontent.com/msyamkumar/cs220-f20-projects/master/p12/techslides-snapshot.html')
with open('capitals.html', 'r') as f:
    contents = f.read()

soup = BeautifulSoup(contents)
table = soup.find_all("table")
rows = soup.find_all("tr")
header = ["country", "capital", "latitude", "longitude"]
html_list = []
for i in range(1, len(rows)):
    country_dict = {}
    data = rows[i].find_all("td")
    if data[header.index("country")].text in countries["country"].unique():
        for j in range(len(header)):
            data_element = data[j].text
            try:
                data_element = float(data_element)
            except ValueError:
                pass
            country_dict[header[j]] = data_element
        html_list.append(country_dict)

output_file = open("my_capitals.json", "w")
json.dump(html_list, output_file)
output_file.close()
open("my_capitals.json", "r").read()
```
